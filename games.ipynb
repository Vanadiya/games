##  Предобработка пропусков

### Пропуски в year_of_release

Удалим строки, где год выпуска не указан. Я решила именно удалить потому что здесь очень широкий диапазон годов (от 1980 до 2015) это большой период и средним значением тут не отделаться (как и медианой). К тому же здесь низкий процент пропусков и эти данные я смело удаляю.

```python
data = data.dropna(subset=['year_of_release'])
```

### Обработка в user_score

В столбце user_score строки с tbd означают, что параметр еще не определён. Эти значения еще не заполнены, поэтому для более удобной работы с данными заменим tbd на 0

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = '-1'
```

Помимо tbd в данном столбце есть пропуски, которые тоже заменим на 0.

```python
data['user_score'] = data['user_score'].fillna(-1)
```

### Обработка в rating

Заполним пропуски в 'rating' пустыми строками. Если рейтинга ESRB нет, то никакими "средними" показателями, или показателями по категории не заполнить.

```python
data['rating'] = data['rating'].fillna('unknown')
```

```python
data['rating'].unique()
```

Я тот самый динозавр, который не играет в компьютерные игры. Но! Википедия <https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board#Устаревшие> подсказала мне, что 'K-A' это старая версия 'E', значит проще подвести под новый рейтинг.

```python
data['rating'] = data['rating'].where(data['rating']!='K-A', 'E') 
data['rating'].unique()
```

## Изменение типа данных

###  year_of_release

```python
data['year_of_release'] = data['year_of_release'].astype('int') 
#изменим тип данных с плавающей точкой на целочисленные
```

###  user_score

```python
data['user_score'] = data['user_score'].astype('float') 
#изменим тип данных с object на float
```

```python
data.info()
```

###  critic_score

```python
data['critic_score'] = data['critic_score'].fillna(-1)
```

```python
data['critic_score'] = data['critic_score'].astype('int') 
#изменим тип данных с плавающей точкой на целочисленные
```

## Продажи по игровым платформам

```python
platforms_sales = data.pivot_table(index = 'platform', values = 'sum_sales', aggfunc='sum').sort_values(ascending=False, by = 'sum_sales')
# создаем сводную таблицу по типам плотформ и сумму дохода от каждой
platforms_sales['platforms']= platforms_sales.index 
# для построения графика дублируем индексы
platforms_sales
```

```python
# импортируем матплотлиб для построения графиков
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(20,8)) # создаем поле для построения графика
sns.barplot(x='platforms',y='sum_sales', data = platforms_sales) # передаем параметры осей
plt.xlabel("Тип игровой платформы") # название оси абсцисс
plt.ylabel("продажи") # название ординат
```

Ярковыраженные пики имеются у шести платформ:PS2, X360, PS3, Wii, DS, PS, что означает их наибольшие продажи. Дальнейшее растределение по годам будет построено для этих лидеров.

```python
# задаем переменную для каждой платформы с помощью срезов предыдущего среза
sales_Wii = selection_platforms.query("platform == 'Wii'").pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum' )
sales_DS = selection_platforms.query("platform == 'DS'").pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum' )
sales_PS2 = selection_platforms.query("platform == 'PS2'").pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum' )
sales_PS3 = selection_platforms.query("platform == 'PS3'").pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum' )
sales_X360 = selection_platforms.query("platform == 'X360'").pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum' )
sales_PS = selection_platforms.query("platform == 'PS'").pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum' )
```

```python
plt.figure(figsize=(15,8)) # создаем поле для построения графика

plt.plot(sales_Wii, color='r', label='Wii') # передаем данные по платформе Wii
plt.plot(sales_DS, color='g', label='DS') # передаем данные по платформе DS
plt.plot(sales_PS2, color='y', label='PS2') # передаем данные по платформе PS2
plt.plot(sales_PS3, color='black', label='PS3') # передаем данные по платформе PS3
plt.plot(sales_X360, color='m', label='X360') # передаем данные по платформе X360
plt.plot(sales_PS, color='indigo', label='PS') # передаем данные по платформе PS
plt.legend() # выводим подписи (легенду)
plt.xlabel("Период") # обозначения абсцисс
plt.ylabel("Продажи") # подпись ординат
```

## Влияние продаж в X360

Для того чтобы посмотреть, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков необходимо построить диаграмму рассеяния и посчитать корреляцию между пользователями и продажами.

```python
X360 = actual_period.query("platform == 'X360'") 
# делаем срез данных по нужной игровой платформе
```

```python
X360.plot(x='user_score' , y='sum_sales' , kind = 'scatter', alpha = 0.9, xlim=(0, 10), color = 'indigo', title = "Зависимость продаж от оценок пользователей")
# диаграмма разброса продаж в зависимости от оценок пользователей

print('Корреляция суммарных продаж от оценок пользователей составляет:', X360['user_score'][X360['user_score']>=0].corr(X360['sum_sales']))
```

Полученное распределение и значение корреляции "говорят" о слабой отрицательной взаимосвязи очков пользователей от суммарных продаж.

```python
X360.plot(x='critic_score' , y='sum_sales' , kind = 'scatter', xlim=(20,100), color = 'indigo',  title = "Зависимость продаж от оценок критиков")
# диаграмма разброса продаж в зависимости от оценок критиков
print('Корреляция суммарных продаж от оценок критиков составляет:', X360['critic_score'][X360['critic_score']>=0].corr(X360['sum_sales']))
```

Зависимость продаж игр на платформе X360 от оценок критиков более вероятна, чем от оценок пользователей, однако значение корреляции имеет среднее значение. Значит оценка критиков немного влияет на продажи.

Созданим функцию add_other для добавления в график оставшиеся жанры и платформы

```python
def add_other(national_df, top_5_df, column_type, sales):
    
    sort_other = national_df.sort_values(by=sales, ascending=False)
    top_5_list = []
    for column in top_5_df[column_type]:
        top_5_list.append(column)

    other_sum = sort_other.query(f'{column_type} not in @@top_5_list')[sales].sum()
    full_fixed_df = top_5_df.append(
        {column_type: 'other', sales: other_sum},
        ignore_index=True
    )
                                 
    return full_fixed_df
```

## Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

**Для проверки данной гипотезы необходимо сформулировать нулевую и альтернативную гипотезы:**

**Нулевая гипотеза H₀: Средние пользовательские рейтинги платформ Xbox One и PC равны;**

**Альтернативная гипотеза H₁: Средние пользовательские рейтинги платформ Xbox One и PC не равны.**

Для проверки гипотез необходимо применить критерий alpha.

Как было сказано в теории- чаще всего alpha принимает значение равное 0,05. При отсутствии у меня какого-либо опыта в данных анализах не вижу смысла, что-то менять в данном параметре.

```python
from scipy import stats as st
import numpy as np

alpha = .05 # критический уровень статистической значимости# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(
    xone_test['user_score'], 
    pc_test['user_score'],
    equal_var = False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

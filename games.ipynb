```python
# –∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –º–∞—Ç–ø–ª–æ—Ç–ª–∏–± –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### –ø–æ –∂–∞–Ω—Ä–∞–º

–°–æ–∑–¥–∞–Ω–∏–º —Ñ—É–Ω–∫—Ü–∏—é add_other –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –≤ –≥—Ä–∞—Ñ–∏–∫ –æ—Å—Ç–∞–≤—à–∏–µ—Å—è –∂–∞–Ω—Ä—ã –∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã

```python
def add_other(national_df, top_5_df, column_type, sales):
    
    sort_other = national_df.sort_values(by=sales, ascending=False)
    top_5_list = []
    for column in top_5_df[column_type]:
        top_5_list.append(column)

    other_sum = sort_other.query(f'{column_type} not in @@top_5_list')[sales].sum()
    full_fixed_df = top_5_df.append(
        {column_type: 'other', sales: other_sum},
        ignore_index=True
    )
                                 
    return full_fixed_df
```

```python
platform_df_jp = add_other(jp_users, top_5_jp_users, 'platform', 'jp_sales')
platform_df_eu = add_other(eu_users, top_5_eu_users, 'platform', 'eu_sales')
platform_df_na = add_other(na_users, top_5_na_users, 'platform', 'na_sales')

genre_df_jp = add_other(jp_genre, top_5_jp_genre, 'genre', 'jp_sales')
genre_df_eu = add_other(eu_genre, top_5_eu_genre, 'genre', 'eu_sales')
genre_df_na = add_other(na_genre, top_5_na_genre, 'genre', 'na_sales')
```

–ö–∞–∫ –±—ã–ª–æ —Å–∫–∞–∑–∞–Ω–æ –≤ —Ç–µ–æ—Ä–∏–∏- —á–∞—â–µ –≤—Å–µ–≥–æ alpha –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ–µ 0,05. –ü—Ä–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–∏ —É –º–µ–Ω—è –∫–∞–∫–æ–≥–æ-–ª–∏–±–æ –æ–ø—ã—Ç–∞ –≤ –¥–∞–Ω–Ω—ã—Ö –∞–Ω–∞–ª–∏–∑–∞—Ö –Ω–µ –≤–∏–∂—É —Å–º—ã—Å–ª–∞, —á—Ç–æ-—Ç–æ –º–µ–Ω—è—Ç—å –≤ –¥–∞–Ω–Ω–æ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–µ.

```python
xone_test = actual_period[(actual_period['platform'] == 'XOne')] 
# —Å–¥–µ–ª–∞–µ–º —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è Xbox –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
print('–°—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –¥–ª—è –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã Xbox —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç:', xone_test['user_score'][xone_test['user_score']>=0].mean())
```

```python
pc_test = actual_period[(actual_period['platform'] == 'PC')] 
# —Å–¥–µ–ª–∞–µ–º —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è PC –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
print('–°—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –¥–ª—è –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã PC —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç:', pc_test['user_score'][pc_test['user_score']>=0].mean())
```

```python
# –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞
xone_test.user_score.min(), pc_test.user_score.min()
```

```python
selection_hypotes = actual_period.query("platform in ('PC','X360')") 
# –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ–≥–æ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –æ—Ç–æ–±—Ä–∞–Ω–Ω—ã—Ö –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º –æ—Ç–±–µ—Ä–µ–º –¥–∞–Ω–Ω—ã–µ 
```

```python
year_X360 = selection_hypotes.query("platform == 'X360'").pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum' )
year_PC = selection_hypotes.query("platform == 'PC'").pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum' )
```

```python
plt.figure(figsize=(15,8)) # —Å–æ–∑–¥–∞–µ–º –ø–æ–ª–µ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞

plt.plot(year_X360, color='m', label='X360') # –ø–µ—Ä–µ–¥–∞–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ X360
plt.plot(year_PC, color='indigo', label='PS') # –ø–µ—Ä–µ–¥–∞–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PS
plt.legend() # –≤—ã–≤–æ–¥–∏–º –ø–æ–¥–ø–∏—Å–∏ (–ª–µ–≥–µ–Ω–¥—É)
plt.xlabel("–ü–µ—Ä–∏–æ–¥") # –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è –∞–±—Å—Ü–∏—Å—Å
plt.ylabel("–ü—Ä–æ–¥–∞–∂–∏") # –ø–æ–¥–ø–∏—Å—å –æ—Ä–¥–∏–Ω–∞—Ç
```

–°—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–º–µ—Ç–Ω–æ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è (–≤–æ–∑–º–æ–∂–Ω–æ, —è –Ω–µ–≤–µ—Ä–Ω–æ —Å–¥–µ–ª–∞–ª–∞ —Å—Ä–µ–∑—ã –¥–∞–Ω–Ω—ã–º —Ç.–µ. –æ–±—ä–µ–º –¥–≤—É—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ—Ü —Ä–∞–∑–Ω—ã–π —Ç.–∫. –≥–æ–¥—ã –≤—ã–ø—É—Å–∫–∞ –∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –¥–≤—É—Ö —ç—Ç–∏—Ö –ø—Ä–∏—Å—Ç–∞–≤–æ–∫ —Ä–∞–∑–Ω—ã–π.)

–ü–æ —Å—Ä–µ–¥–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏—è–º –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π —è —Å–º–µ–ª–æ –º–æ–≥—É –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –ø–æ–¥—Ç–≤–µ—Ä–¥–∏—Ç—Å—è.

```python
from scipy import stats as st
import numpy as np

alpha = .05 # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏# –µ—Å–ª–∏ p-value –æ–∫–∞–∂–µ—Ç—Å—è –º–µ–Ω—å—à–µ –Ω–µ–≥–æ - –æ—Ç–≤–µ—Ä–≥–Ω–µ–º –≥–∏–ø–æ—Ç–µ–∑—É

results = st.ttest_ind(
    xone_test['user_score'], 
    pc_test['user_score'],
    equal_var = False)
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)
if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É") 
```

```python
# –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ 2
xone_test.user_score.min(), pc_test.user_score.min()
```

## –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports —Ä–∞–∑–Ω—ã–µ.

**–î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –¥–∞–Ω–Ω–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –Ω—É–ª–µ–≤—É—é –∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É—é –≥–∏–ø–æ—Ç–µ–∑—ã:**

**–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ H‚ÇÄ: –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports —Ä–∞–≤–Ω—ã;**

**–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ H‚ÇÅ: –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports —Ä–∞–∑–Ω—ã–µ –Ω–µ —Ä–∞–≤–Ω—ã.**

```python
action_test = actual_period[(actual_period['genre'] == 'Action')] 
# —Å–¥–µ–ª–∞–µ–º —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —ç–∫—à–µ–Ω –∂–∞–Ω—Ä–∞
print('–°—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –¥–ª—è –∂–∞–Ω—Ä–∞ Action —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç:',action_test['user_score'][action_test['user_score']>=0].mean())
```

```python
sports_test = actual_period[(actual_period['genre'] == 'Sports')] 
# —Å–¥–µ–ª–∞–µ–º —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Å–ø–æ—Ä—Ç–∏–≤–Ω–æ–≥–æ –∂–∞–Ω—Ä–∞ 
print('–°—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –¥–ª—è Sports –∂–∞–Ω—Ä–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç:', sports_test['user_score'][sports_test['user_score']>=0].mean())
```

```python
alpha = .05 # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏# –µ—Å–ª–∏ p-value –æ–∫–∞–∂–µ—Ç—Å—è –º–µ–Ω—å—à–µ –Ω–µ–≥–æ - –æ—Ç–≤–µ—Ä–≥–Ω–µ–º –≥–∏–ø–æ—Ç–µ–∑—É

results = st.ttest_ind(
    action_test['user_score'], 
    sports_test['user_score'],
    equal_var = False)
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)
if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É") 
```

–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –ø–æ–¥—Ç–≤–µ—Ä–¥–∏–ª–∞—Å—å.

6) –ö–∞–∂–¥–∞—è –∏–≥—Ä–æ–≤–∞—è –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ –∏–º–µ–µ—Ç —Å–≤–æ–π —Ä–µ–π—Ç–∏–Ω–≥, –∫–∞–∫ —Å—Ä–µ–¥–∏ –∏–≥—Ä–æ–∫–æ–≤ —Ç–∞–∫ –∏ —Å—Ä–µ–¥–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤. (–Ø —Å—á–∏—Ç–∞—é, —á—Ç–æ —Å–ª–µ–¥—É–µ—Ç –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Å—Ä–µ–¥–Ω–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º —Å—Ä–µ–¥–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞ –¥–ª—è —É—Ç–æ—á–Ω–µ–Ω–∏—è –¥–∞–Ω–Ω–æ–≥–æ –ø—É–Ω–∫—Ç–∞. –ü–æ—Ç–æ–º—É —á—Ç–æ —É —Ä–µ–≥–∏–æ–Ω–æ–≤ –µ—Å—Ç—å —è—Ä–∫–∏–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏—è—Ö, –∫–∞–∫ –ø–æ –∂–∞–Ω—Ä–∞–º –∏–≥—Ä—ã, —Ç–∞–∫ –∏ –ø–æ –∏–≥—Ä–æ–≤—ã–º –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º.)
